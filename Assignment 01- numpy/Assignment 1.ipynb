##### Write a program that will find numbers between 2000 and 3000 that are divisible by 7 but not by 5.
##### Print all these numbers. Try to find a way to print the numbers in one line, separated by a comma. 

```python
 # if include 3000 so its range 2000-3001 if not change 3001 to 3000
```

```python
def find_div_numbers():
    for i in range(2000,3001):
         if((i%7==0)&(i%5!=0)):
            print(i,end=",")        
```
